```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python
data_file_name = 'data-v18-21-29-17-06-21'
df_basic = pd.read_csv('data/'+ data_file_name +'_cleaned.csv', encoding= 'ISO-8859-8')
df_basic.info()
```

```python
import matplotlib.pyplot as plt
display(HTML('<h1 style="font-size:24px;">Rising prices in each city</h1>'))
cities = df_basic.City.unique()
for city in cities:
    display(HTML('<h2 style="font-size:21px;">'+city+'</h2>'))
    sns.lineplot(x="DealYear", y="PricePerSq", data= df_basic.loc[df_basic['City'] == city])
    plt.show()
```

```python
display(HTML('<h1 style="font-size:24px;">General prices rising - each city</h1>'))

plt.figure(figsize=(12, 7))
sns.lineplot(x="DealYear", y="PricePerSq", data= df_basic,hue="City")

```

```python
display(HTML('<h1 style="font-size:24px;">General prices rising - all cities</h1>'))
plt.figure(figsize=(12, 7))

sns.lineplot(x="DealYear", y="PricePerSq", data= df_basic)
```

```python
display(HTML('<h1 style="font-size:24px;">prices comparison  between each city between 2010 - 2020</h1>'))

df_prices_2010 = df_basic.loc[df_basic['DealYear'] == 2010][['City','DealYear','PricePerSq']]
df_prices_2020 = df_basic.loc[df_basic['DealYear'] == 2020][['City','DealYear', 'PricePerSq']]
df_prices_2015 = df_basic.loc[df_basic['DealYear'] == 2015][['City','DealYear', 'PricePerSq']]

df_prices = df_prices_2010.copy()
df_prices = pd.concat([df_prices,df_prices_2015],ignore_index=True)
df_prices = pd.concat([df_prices,df_prices_2020],ignore_index=True)
ax = plt.figure(figsize=(12, 12))
clrs = ['#cc8963','#5F9E6E','#857AAB']
ax = sns.barplot(x="City", y="PricePerSq",hue='DealYear',data=df_prices,palette=clrs)
```

```python
import matplotlib.pyplot as plt
display(HTML('<h1 style="font-size:24px;">Rising prices in each city</h1>'))
cities = df_basic.City.unique()
df_basic['DealYear'] = df_basic['DealYear'].astype(int)
for city in cities:
    display(HTML('<h2 style="font-size:21px;">'+city+'</h2>'))
    ax = plt.figure(figsize=(7, 7))
    sns.barplot(x="DealYear", y="PricePerSq",data= df_basic.loc[df_basic['City'] == city],palette=clrs)
    plt.show()
```
